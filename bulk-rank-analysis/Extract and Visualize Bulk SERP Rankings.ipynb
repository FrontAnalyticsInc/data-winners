# Extract and Visualize Bulk SERP Rankings

## Competitive Analysis


Inputs:

- a set of important keywords
- a list of competitors' domains

Objective:

- group keywords into semantically similar topics
- Extract the best rank for your site and all the competitors
- Use ChatGPT code interpreter to visualize the current rank for all the sites 

Use https://app.contentcurator.com/playbook/serp-similarity/ 


## Prompt for ChatGPT code interpreter:

```
bar chart in python of the following data showing average rank for salomon and its competitors. set the color of the bar to the brand's primary color:

salomon	50
adidas	42.5
nike	41
newbalane	27
brooks	27
```

Here is the result:

````python
import matplotlib.pyplot as plt

# Data
competitors = ['Salomon', 'Adidas', 'Nike', 'New Balance', 'Brooks']
ranks = [50, 42.5, 41, 27, 27]

# Brand colors dictionary
brand_colors = {
    'Salomon': '#FF0000',        # Red
    'Adidas': '#0000FF',         # Blue
    'Nike': '#FFA500',           # Orange
    'New Balance': '#008000',    # Green
    'Brooks': '#FFC0CB'          # Pink
}

# Create bar chart with brand colors
bars = plt.bar(competitors, ranks)

# Set the bar colors using brand colors dictionary
for competitor, bar in zip(competitors, bars):
    bar.set_color(brand_colors[competitor])

# Add labels and title
plt.xlabel('Competitor')
plt.ylabel('Average Rank')
plt.title('Average Rank for Salomon and Competitors')


# Display the chart
plt.show()
````
